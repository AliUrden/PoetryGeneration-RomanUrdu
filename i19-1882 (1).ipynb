## Below is the Function which converts urdu to roman using above dictionary 

```python
def urdu2roman(word):
    #looking at the string character by character and based on character position in word returning english couterpart
    rw=''
    for char in word:
        #in case if the character is an digit return as it is
        if char.isdigit():    
            rw = rw + char
            
        #in case if the character is an English alphabet return as it is
        elif (ord(char) >= 65) and (ord(char) <= 90):    
            rw = rw + char
            
        #in case if the character is an English alphabet return as it is
        elif (ord(char) >= 97) and (ord(char) <= 122):
            rw = rw + char
            
        # in case the character is from urdu                             
        elif word.startswith(char):
            #print(urdu_dict[ord(char)]['start'])
            rw = rw + (urdu_to_roman_dict[hex(ord(char))]['start'])
        elif word.endswith(char):
            #print(urdu_dict[ord(char)]['end'])
            rw = rw + (urdu_to_roman_dict[hex(ord(char))]['end'])
        else:
            #print(urdu_dict[ord(char)]['mid'])
            rw = rw + (urdu_to_roman_dict[hex(ord(char))]['mid'])
    return(rw)
```

##  FIle Reading & Preprocessing
Reading file + removing Punctuations 

```python
#reding the file
word_tokens = []
punc = '''!()%\n٪-;۔،:\n\/'"\,“./؟_ء'''

f = codecs.open('poetry.txt', encoding = "utf_8")
for line in f:
    #converting to spacy.tokens so that word tokeniztion can be achieved easily
    sen = unlp(line)
    for w in sen:
        #converting the word back to 'str'
        w = str(w)
        #preprocessing (removing punctuations)
        for c in w:
            if c in punc:
                w = w.replace(c,'')
                
        if w != "\n" and w != "\r" and w != " " and w != '' and w != " \r" and w !='‘' and w !='’':
            word_tokens.append(w)
f.close()
```

```python
def next_probable_rbi(w):
    for i in rev_bi_grams_freq:
        if i[0].split(' ')[0] == w: #checking the probability of (w2) given w1 
            nw = i[0].split(' ')[1]
            return(nw)
            #print(i)
```

```python
verse = ''
stanzaa3 = []
for stanza in range(7):
    for verses in range(2):
        wo = random.randint(6,7) # number of words inside a varse
        last_word_ind = random.randint(0,(len(word_tokens)-1)) #randomly generating index of first word
        last_word = word_tokens[last_word_ind] # geting word at index
        verse = last_word
        for word in range(wo):
            #print(last_word)
            next_word = next_probable_rbi(last_word)
            verse = verse + ' ' + next_word 
            last_word = next_word
        #reversing the string as it tells about prior    
        lis = []
        lis = verse.split(' ')
        lis.reverse()
        verse = ' '.join(lis)
        print(str(verse))
        stanzaa3.append(verse)
        verse = ''
    print('\n')


            
```
